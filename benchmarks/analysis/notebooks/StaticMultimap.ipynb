```python
# !pip install pandas
# !pip install matplotlib


# Import libraries
import pandas as pd
import matplotlib

# Define the data path
datafile='../data/static-multimap-data.csv'
```

# Import Data

```python
# Read csv file
rawdf = pd.read_csv(datafile)

# Filter out unrelated data
perfdf = rawdf[rawdf["Key"] == rawdf["Value"]].reset_index(drop=True)
perfdf.drop(columns=['Device', 'Device Name', 'Skipped'], inplace=True)

# Drop NANs
perfdf.dropna(inplace=True)
```

# Comput Performance

```python
perfdf['Performance'] = perfdf['NumInputs'] / perfdf['GPU Time (sec)']
#display(perfdf)
```

# Visualize Performance

```python
perfdf.plot(x="Occupancy", y="Performance");
```
